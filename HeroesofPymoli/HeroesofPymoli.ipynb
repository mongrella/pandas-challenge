```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/HeroesOfPymoli_Resources_purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load, encoding="utf-8")
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#use the len function to get unique SN
total_players = len(purchase_data["SN"].value_counts())

players = pd.DataFrame({"Total Players":[total_players]})
players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Unique items = len, unique, item ID
#average price = mean, price
#number of purchases = value counts, purchase ID
#total revenue = sum, price

unique_items = len((purchase_data["Item ID"]).unique())
average_price = (purchase_data["Price"]).mean()
total_purchases = (purchase_data["Purchase ID"]).count()
total_revenue = (purchase_data["Price"]).sum()

# Create summary data frame to hold results
summary_df = pd.DataFrame({"Number of Unique Items":[unique_items],
                           "Average Price":[average_price], 
                           "Number of Purchases": [total_purchases], 
                           "Total Revenue": [total_revenue]})
#Curreny formatting code taken from https://pbpython.com/styling-pandas.html
summary_df.style.format({'Average Price':'${0:,.2f}',
                         'Total Revenue':'${0:,.2f}'})
```
